## Exercise 3: Non-linear Regression with gradient descent

The gradient descent is a technique to find the minimum of a function.

We aim to use Gradient Descent to minimize the error of the regression. We aim to find the coefficients $\boldsymbol{\beta}$ that minimize the reconstruction error of a function $f$:

 - Non linear function : $f(x)$
 - Regression: $\hat{\beta} = \textit{argmin}\{ || y - f(x) ||^2 \}$

Therefore, we need to define $y=y(\boldsymbol{\beta},x)$
